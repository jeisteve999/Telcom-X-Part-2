```python
y_smote.value_counts()
```

# Given the imbalance between active customers and those who canceled, balancing techniques were applied to improve the predictive capacity of the models. In particular, SMOTE was used to generate synthetic examples of the minority class, avoiding the loss of information associated with undersampling and reducing the risk of overfitting present in traditional oversampling.

# Training the model without SMOTE

```python
from sklearn.model_selection import train_test_split

x = telcomx_2.drop('churn',axis =1)
y = telcomx_2['churn']

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42, stratify = y)
```

```python
# model without base NOT SMOTE

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

model_base = LogisticRegression(max_iter = 1000)
model_base.fit(x_train, y_train)

y_ped_base = model_base.predict(x_test)

print("without SMOTE")
print(classification_report(y_test, y_ped_base))
```

```python
#apply smote only train

from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)

x_train_smote, y_train_smote = smote.fit_resample(x_train, y_train)
```

```python
# Model with SMOTE
model_smote = LogisticRegression(max_iter = 1000)
model_smote.fit(x_train_smote, y_train_smote)
y_pred_smote = model_smote.predict(x_test)

print("with SMOTE")
print(classification_report(y_test, y_pred_smote))
```

# When comparing models with and without balancing, it is observed that applying SMOTE improves the model's ability to identify customers who cancel service, increasing recall of the minority class without significantly compromising overall accuracy. This demonstrates that balancing contributes to a more useful model for churn prediction.

# The proportion of customers who remain active is significantly higher than that of those who canceled their service. This indicates a moderate imbalance between the classes, which could affect the performance of predictive models, as these could be biased toward the majority class if appropriate metrics or techniques are not considered.

# Since the dataset contains numerical variables at different scales, such as tenure and monthly, normalization or standardization techniques are necessary when using scale-sensitive models like Logistic Regression, KNN, SVM, or Neural Networks. In contrast, tree-based models like Decision Tree, Random Forest, or XGBoost do not require this preprocessing, as their operation is independent of distances between variables.

#Predictive modeling


```python
X = telcomx_2.drop('churn', axis=1)
y = telcomx_2['churn']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,      # 20% prueba
    random_state=42,    # reproducibilidad
    stratify=y          # mantiene proporción de churn
)
```

```python
print("Train shape:", X_train.shape)
print("Test shape:", X_test.shape)
```

```python
print("Train churn ratio:\n", y_train.value_counts(normalize=True))
print("Test churn ratio:\n", y_test.value_counts(normalize=True))
```

# Training models

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```python
# Logistic Regression

log_model = LogisticRegression(max_iter=1000)
log_model.fit(X_train, y_train)

y_pred_log = log_model.predict(X_test)
```

```python
# Random forest

rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)

y_pred_rf = rf_model.predict(X_test)
```

```python
# smv

svm_model = SVC()
svm_model.fit(X_train, y_train)

y_pred_svm = svm_model.predict(X_test)
```

```python
# evaluating models

def evaluar_modelo(y_true, y_pred, nombre):

    print(f"\n {nombre}")
    print("Accuracy:", accuracy_score(y_true, y_pred))
    print("Precision:", precision_score(y_true, y_pred))
    print("Recall:", recall_score(y_true, y_pred))
    print("F1 Score:", f1_score(y_true, y_pred))

    print("\nConfusion Matrix:")
    print(confusion_matrix(y_true, y_pred))
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
```

```python
evaluar_modelo(y_test, y_pred_log, "Logistic Regression")
evaluar_modelo(y_test, y_pred_rf, "Random Forest")
evaluar_modelo(y_test, y_pred_svm, "SVM")
```

```python
print("RF Train:", rf_model.score(X_train, y_train))
print("RF Test:", rf_model.score(X_test, y_test))
```

Logistic Regression showed signs of overfitting, achieving very high performance on training data (0.99) but lower performance on test data (0.78). This suggests the model may be learning noise or overly specific patterns instead of generalizable trends. The large number of encoded categorical variables may be contributing to this behavior.

Logistic Regression achieved the best F1-score among the tested models, indicating the strongest balance between precision and recall when predicting churn. While Random Forest and SVM may perform well in accuracy, Logistic Regression was more effective at correctly identifying customers at risk of cancellation. This makes it the most suitable model for churn prediction, where detecting potential losses is more important than overall correctness.

# The regularized Logistic Regression model reduced overfitting by achieving more consistent performance between training and test data. However, this came at the cost of a slightly lower F1-score, driven mainly by a drop in recall. While the model became more stable, it also became less sensitive in detecting customers at risk of churn.
